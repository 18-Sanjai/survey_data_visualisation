```python
import pandas as pd

df = pd.read_csv("C:\\Users\\SANJAIKRISHNAN\\Downloads\\Starbucks satisfactory survey encode cleaned.csv")
```

```python
display(df.head())
display(df.columns)
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for column in df.columns:
    if column in ['Id', 'loyal']: 
        continue
    if df[column].nunique() < 10 or df[column].dtype == 'object':
        plt.figure(figsize=(8, 6))
        sns.countplot(x=column, data=df)
        plt.title(f'Distribution of {column}')
        plt.xlabel(column)
        plt.ylabel('Count')
        plt.show()
        plt.close()
    elif df[column].nunique() <= 5: 
        plt.figure(figsize=(8, 8))
        df[column].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
        plt.title(f'Proportion of {column}')
        plt.ylabel('') # Hide the default y-label
        plt.show()
        plt.close()


plt.figure(figsize=(12, 10))
correlation_matrix = df.drop(columns=['Id']).corr()
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
plt.title('Correlation Matrix of Survey Responses')
plt.show()
plt.close()
```
